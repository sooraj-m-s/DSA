# Stone Paper Scissors

```python
import random


class StonePaperScissors:
    def __init__(self, limit):
        self.choice = 0
        self.system = 0
        self.player = 0
        self.limit = limit
        self.round = 1
        self.first = True

    def system_choice(self):
        return random.randint(1, 3)

    def get_value(self):
        try:
            while self.round <= self.limit:
                print(f'\n\nRound: {self.round}\n1) Stone\n2) Paper\n3) Scissors\n4) Exit\nEnter you choice: ')
                choice = int(input())
                while choice < 1 or choice > 4:
                    choice = int(input('Invalid input, enter again: '))
                if choice == 4:
                    print('Exited...!')
                    return
                else:
                    self.round += 1
                    self.play(choice, self.system_choice())
        except ValueError as v:
            print('Invalid input. Program exited...!')
        else:
            if self.system > self.player:
                print('Unfortunatly you lose the game. Please come back again...!')
                return
            elif self.system == self.player:
                print('Bad luck, game is a tie...!')
                return
            else:
                print('Congrats, you won the game...!')
                return

    def play(self, choice, system_choice):
        if choice == 1 and system_choice == 1:
            print(f'\nYour choice: Stone\nSystem choice: Stone\nTie...\nScore:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 1 and system_choice == 2:
            print(f'\nYour choice: Stone\nSystem choice: Paper\nSystem won...')
            self.system += 1
            print(f'Score:-\nYour score: {self.player}\nSystem choice: {self.system}')
            return
        elif choice == 1 and system_choice == 3:
            print(f'\nYour choice: Stone\nSystem choice: Scissors\nYou won...')
            self.player += 1
            print(f'Score:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 2 and system_choice == 1:
            print(f'\nYour choice: Paper\nSystem choice: Stone\nYou won...')
            self.player += 1
            print(f'Score:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 2 and system_choice == 2:
            print(f'\nYour choice: Paper\nSystem choice: Paper\nTie...\nScore:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 2 and system_choice == 3:
            print(f'\nYour choice: Paper\nSystem choice: Scissors\nSystem won...')
            self.system += 1
            print(f'Score:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 3 and system_choice == 1:
            print(f'\nYour choice: Scissors\nSystem choice: Stone\nSystem won...')
            self.system += 1
            print(f'Score:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 3 and system_choice == 2:
            print(f'\nYour choice: Scissors\nSystem choice: Paper\nYou won...')
            self.player += 1
            print(f'Score:-\nYou: {self.player}\nSystem: {self.system}')
            return
        elif choice == 3 and system_choice == 3:
            print(f'\nYour choice: Scissors\nSystem choice: Scissors\nTie...\nScore:-\nYou: {self.player}\nSystem: {self.system}')
            return


user_choice = input('Welcome to the game Stone Paper Scissors. Enter the round limit to continue: ')
limit = None
try:
    limit = int(user_choice)
    if limit < 1:
        raise ValueError
except ValueError:
    if limit is None:
        print(f'{user_choice} invalid input. Program exited...!')
    else:
        print(f'Invalid round limit {limit}. Round must be a positive integer.')
else:
    game = StonePaperScissors(limit)
    game.get_value()

```
